```python
from wordcloud import WordCloud

# Select the data you want to use to create the word cloud
query = "SELECT artist_name FROM artists.artists_dataset"
cursor.execute(query)

# Fetch the data from the database
data = cursor.fetchall()

# Convert the data to a single string
text = " ".join(row[0] for row in data)

# Create a WordCloud object
wordcloud = WordCloud().generate(text)

# Display the word cloud
import matplotlib.pyplot as plt
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()

# Close the cursor and the connection
# cursor.close()
# conn.close()
```

```python
query = '''SELECT artist_comments, LENGTH(artist_comments), LENGTH(TRIM(artist_comments)), ASCII(artist_comments), CONVERT(artist_comments, UNSIGNED),
                  artist_favorites, LENGTH(artist_favorites), LENGTH(TRIM(artist_favorites)), ASCII(artist_favorites), CONVERT(artist_favorites, UNSIGNED)
           FROM artists.artists_dataset 
           ORDER BY CONVERT(artist_comments, UNSIGNED), CONVERT(artist_favorites, UNSIGNED)'''
cursor.execute(query)
data_plot = cursor.fetchall()

# cursor.close()

x = [x[4] for x in data_plot]
y = [y[9] for y in data_plot]

# plt.figure(figsize=(20, 20))
plt.scatter(x,y)
plt.title('Average Comments by Different Favorites',fontsize=12)
plt.xlabel('comments',fontsize=10)
plt.ylabel('favorites',fontsize=10)
plt.show()
```
